# Plot sigmoid
iPython notebook with the sigmoid function and the plot (incl. derivative).

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-8, 8, 0.1)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

First derivative of sigmoid


```python

d1_f = (np.exp(-x) / (1 + np.exp(-x))**2)
plt.plot(x, d1_f)
plt.xlabel('x')
plt.ylabel('d/dx sigmoid\'(x)')
plt.show()
```

```python
d2_f = -(((np.exp(x) - 1)*np.exp(x)) / (1 + np.exp(x))**3)
plt.plot(x, d2_f)
plt.xlabel('x')
plt.ylabel('d/dx sigmoid\'\'(x)')
plt.show()
```

```python
# Exercise 2. c) Zeta

# Compute the limits with z -> +/- infinity
# of -ln(sigma(-z)), where sigma is the sigmoid function


def zeta(z):
    sig = sigmoid(-z)
    return -np.round(np.log(sig))
```

```python
x = np.zeros((100,1))
for n in range(0,100):
    x[n]= n*10
x1 = np.array([0,1,10,100,700])
x2 = np.array([0,-1,-10,-100,-700])

y1 = zeta(x1)
y2 = zeta(x2)
l1, = plt.plot(x1,y1)
l2, = plt.plot(x2,y2)
plt.legend([l1,l2],["z going to plus infinity","z going to minus infinity"])
plt.ylabel('Y')
plt.xlabel('X')
plt.show()
```
