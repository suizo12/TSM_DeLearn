# Plot sigmoid
iPython notebook with the sigmoid function and the plot (incl. derivative).

```python
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-8, 8, 0.1)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

First derivative of sigmoid


```python

d1_f = -(np.exp(-x) / (1 + np.exp(-x))**2)
plt.plot(x, d1_f)
plt.xlabel('x')
plt.ylabel('d/dx sigmoid\'(x)')
plt.show()
```

```python
d2_f = -(((np.exp(x) - 1)*np.exp(x)) / (1 + np.exp(x))**3)
plt.plot(x, d2_f)
plt.xlabel('x')
plt.ylabel('d/dx sigmoid\'\'(x)')
plt.show()
```
